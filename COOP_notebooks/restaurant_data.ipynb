## **Cleaning the Data**

### **Finding, Deleting Duplicates**

```python
# Finding Duplicates
duplicates = df[df.duplicated()]
```

```python
# Keeps the first instance and drops any remaining duplicates

df_clean = df.drop_duplicates(keep='first')
```

```python
df_clean.shape
```

### **Dropping Rows with Null Values**

```python
# Checks for any null values

df_clean.isna()
```

```python
# Drops rows with null values

df_clean = df_clean.dropna(axis='rows')
```

### **Subsetting Columns**

```python
# New dataframe with select columns

df_clean = df_clean[['CAMIS', 'DBA', 'BORO', 'ZIPCODE', 'CUISINE DESCRIPTION', 'INSPECTION DATE', 'ACTION', 'VIOLATION CODE', 'VIOLATION DESCRIPTION', 'CRITICAL FLAG', 'INSPECTION TYPE', 'Latitude', 'Longitude']]
```
