## **Example DataFrames**

```python
import pandas as pd
# df1 = pd.DataFrame({"city":["new york", "chicago", "orlando"],
#                    "temperature":[21,14,35]})
# df1

df1 = pd.DataFrame({"city":["new york", "chicago", "orlando", "baltimore"],
                   "temperature":[21,14,35,32]})
df1
```

```python
# df2 = pd.DataFrame({
#     "city":["chicago","new york", "orlando"],
#                     "humidity":[65,68,75]
#                     })
# df2

df2 = pd.DataFrame({
    "city":["chicago","new york", "san francisco"],
                    "humidity":[65,68,75]
                    })
df2
```

## **Inner Join (Default Merge, how="inner")**

```python
# df3 = pd.merge(df1,df2, on="city")
# df3

df3 = pd.merge(df1, df2, on="city", how="inner")
df3

# Only joins on the similarities between the two dataframes and excludes everything else (think of the center of a venn diagram)
# by default, how is "inner"
```

## **Outer Join (how="outer")**

```python
df4 = pd.merge(df1, df2, on="city", how="outer")
df4

# includes everything regardless of similarities (a union)
```

## **Left Join (how="left")**

```python
df5 = pd.merge(df1, df2, on="city", how="left")
df5

# Everything in the left dataframe, including where the similarities between the two dataframes are, and excludes everything else (san francisco)
```

## **Right Join (how="right")**

```python
df6 = pd.merge(df1, df2, on="city", how="right")
df6

# Everything in the left dataframe, including where the similarities between the two dataframes are, and excludes everything else (orlando, baltimore)
```

## **Indicator Flag**

```python
df7 = pd.merge(df1, df2, on="city", how="outer", indicator=True) #Boolean, NOT a string
df7

#  Let's you know what data came from which dataframe
```

```python
df1 = pd.DataFrame({
    "city":["new york", "chicago", "orlando", "baltimore"],
    "temperature":[21,14,35,32],
    "humidity": [65,68,71,75]})
df1
```

```python
df2 = pd.DataFrame({
    "city":["chicago","new york", "san francisco"],
    "temperature": [21,14,35],
    "humidity":[65,68,75]})
df2
```

## **Custom Suffixes for Columns**

```python
df8 = pd.merge(df1, df2, on="city", suffixes=('_left', '_right'))
df8
```
