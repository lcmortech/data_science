## **Example DataFrames**

```python
import pandas as pd
# df1 = pd.DataFrame({"city":["new york", "chicago", "orlando"],
#                    "temperature":[21,14,35]})
# df1

df1 = pd.DataFrame({"city":["new york", "chicago", "orlando", "baltimore"],
                   "temperature":[21,14,35,32]})
df1
```

```python
# df2 = pd.DataFrame({
#     "city":["chicago","new york", "orlando"],
#                     "humidity":[65,68,75]
#                     })
# df2

df2 = pd.DataFrame({
    "city":["chicago","new york", "san francisco"],
                    "humidity":[65,68,75]
                    })
df2
```

## **Inner Join (Default Merge, how="inner")**

```python
# df3 = pd.merge(df1,df2, on="city")
# df3

df3 = pd.merge(df1, df2, on="city", how="inner")
df3

# Only joins on the similarities between the two dataframes and excludes everything else (think of the center of a venn diagram)
# by default, how is "inner"
```

## **Outer Join (how="outer")**

```python
df4 = pd.merge(df1, df2, on="city", how="outer")
df4

# includes everything regardless of similarities (a union)
```

## **Left Join (how="left")**

```python
df5 = pd.merge(df1, df2, on="city", how="left")
df5

# Includes everything in left dataframe, and where the similarities between the two dataframes are, and excludes everything else (san francisco)
```
